# Pie Charts
---
Pie Charts shows the proportion that each category contributes to the whole.<br>
<br>**Note:** Pie charts may not accurately visualize small differences, so in such cases, it is advisable to use bar charts instead

```python
# import packages
import matplotlib.pyplot as plt
import pandas as pd
import numpy
```

## A simple Pie Plot
`plt.pie(slices,labels)` is the code to plot a pie chart.

```python
# data to plot
slices=[60,40]
labels=['Sixty','Forty']

plt.pie(x=slices,labels=labels)
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

### Wedge Properties
The properties of the wedges can be changes using the `wedgeprop` argument.<br>
Properties like 'linewidth', 'edgecolor' can be passed, eg- `wedgeprops={"linewidth": 1, "edgecolor": "white"}`<br>
Checkout matplotlib's [Pie Plot official documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.pie.html) for more. <br>
The color of each wedge can also be manually changes by passing in a list of color.

```python
# data to plot
slices=[40,30,20]
labels=['Forty','Thirty','Twenty']
colors=['#CCCCFF','#5BCEFA','#F5A9B8']

plt.pie(x=slices,labels=labels,colors=colors,wedgeprops={"linewidth": 1, "edgecolor": "white"})
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

### Plot Language Popularity

```python
# Language Popularity
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']
```

```python
plt.pie(x=slices,labels=labels,wedgeprops={"linewidth": 0.5, "edgecolor": "black"})
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

As mentioned in the beginning, here there are way too many categories and the wedges are not distinguishable enough. This is a good example where using Bar Chart would be better.<br>
Usually Pie chart is used only when there are under 5 categories. 

To fix the above plot let's just use the top 5 languages

```python
plt.pie(x=slices[:5],labels=labels[:5],wedgeprops={"linewidth": 0.5, "edgecolor": "black"})
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

For this specific data this still doesn't help as the plot still doesn't have distinguisable wedges, but this is to demonstrate the general idea behind it.

### Explode wedges
When we want to highlight particular wedges, we use the **`explode`** argument.<br>
It is a list of floats representing each wedge and the values range between 1-0. Where **'0'** means no exploded wedge and a value closer to **'1'** means that much percent of the radius exploded for that wedge.<br>
Eg- Say a wedge has **'0.1'** as it's value, that would mean it's been exploded **10%** the size of radius away from the center.

```python
explode=[0.1,0,0,0,0]
```

```python
plt.pie(x=slices[:5],
        labels=labels[:5],
        explode=explode,
        wedgeprops={"linewidth": 0.5, "edgecolor": "black"})
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

### Pie chart customizations

**Shadows**: Shadow argument is a boolean argument which when set to `True`, shows a shadow for the chart. This is just for visual aesthetics and commonly used with explode as it gives a more 3d look, however it's purely aesthetics and personal preference of when to use it and how.

**Startangle**: This is used to rotate the chart by a specified degree from its original default orientation.

```python
plt.pie(x=slices[:5],
        labels=labels[:5],
        explode=explode,
        shadow=True,
        startangle=70,
        wedgeprops={"linewidth": 0.5, "edgecolor": "black"})
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

### Show Percentage in Pie Chart
When we want to show the percentage value of each wedge we use the **`autopct`** argument which takes a format string as input. Eg- **`%1.1f%%`**<br><br>
Here is a Chat-GPT 3.5 generated explanation:<br><br>
The format string %1.1f%% is used in programming, particularly in languages like Python, to format and display numbers as strings in a specific way. Let's break down each part:

**`1.`** `%`: This is the modulus operator used in programming. However, when used in a format string, it serves as a placeholder for values to be inserted.

**`2.`** **`1.1f`**: This is a specification for formatting floating-point numbers (numbers with decimal points). Let's break it down further:
- The first **`1`** represents the minimum width of the entire number, including the decimal point and any leading digits. In this case, it's set to 1, meaning there will be at least one digit before the decimal point.
- The **`.`** is a literal dot character that separates the width and precision specifications.
- The second **`1`** after the dot specifies the number of digits to display after the decimal point.
- f indicates that the number is a floating-point number.

**`3.`** `%%`: The final `%%` is a literal percentage sign. The double percentage signs are used to escape and display a single percentage sign in the resulting string.

```python
plt.pie(x=slices[:5],
        labels=labels[:5],
        explode=explode,
        shadow=True,
        startangle=70,
        autopct='%1.1f%%',
        wedgeprops={"linewidth": 0.5, "edgecolor": "black"})
plt.title("Simple Pie Plot")
plt.tight_layout()
plt.show()
```

That's it for Pie Chart it is good to get an overview of the data however as mentioned multiple times when we want to get proper insight from the data a Bar Chart is almost always better.
