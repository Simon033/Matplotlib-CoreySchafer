# Subplots
If we want to plot multiple plots or if we want to create plot in an OOP way, we can create using a subplot method.<br><br>
When we create a single plot using plt it directly creates a Pyplot object in a stateful manner using methods without explicitly creating an object, using subplots make the plot more manageable by creating instance. Some people prefer subplots even when just creating one plot.

```python
# import packages
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# import data
data = pd.read_csv('./data/data10.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']
```

### Simple Plot

```python
# plot python salary
plt.plot(ages, py_salaries, label='Python')
# plot javascript salary
plt.plot(ages, js_salaries, label='JavaScript')
# plot all dev salary
plt.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

plt.legend()

plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.tight_layout()
plt.show()
```

The above is stateful, because a current state maintatins memory of what current figure and axes we are working with.

**Figure** is a container that shows and holds the plots within it.<br>
**Axes** are the actual plots.

### Simple Subplot method
Lets recreate the same plot using subplot method.<br>
To create that we will use `fig, ax=plt.subplots()`, where `fig`refers to the figure object and `ax` refers to the plots and we have to use `ax` instead of `plt` to reference and create/change the plot.<br>
When using this method we can pass in the number of rows and columns we want however if nothing it passsed, it defaults to a 1x1 single plot.

```python
fig,ax=plt.subplots()

# plot python salary
ax.plot(ages, py_salaries, label='Python')
# plot javascript salary
ax.plot(ages, js_salaries, label='JavaScript')
# plot all dev salary
ax.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax.legend()

# to set title and lables we have to use set_.. instead
ax.set_title('Median Salary (USD) by Age')
ax.set_xlabel('Ages')
ax.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

### Creating Multiple Plots
Lets create a plot of Python and Javascript and another seperate plot of All Dev.<br>
To do that we just have to pass a the `ncols` and `nrows` argumetns inside `plt.subplots()`

First lets understand how the method returns the axes. It returns a numpy array or an array of array, with the axes objects inside.<br>

```python
fig,ax=plt.subplots(nrows=2,ncols=1)
print(ax)
print(type(ax))
print(type(ax[1]))
plt.close()
print('---')
fig,ax=plt.subplots(nrows=2,ncols=3)
print(ax)
print(type(ax))
print(type(ax[1]))
plt.close()
```

When creating seperate plots we can use **unpacking** to get seperate axes and specify them when setting titles/labels

```python
fig,(ax1,ax2)=plt.subplots(nrows=2)

# plot all dev salary
ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
# plot python salary
ax2.plot(ages, py_salaries, label='Python')
# plot javascript salary
ax2.plot(ages, js_salaries, label='JavaScript')


ax1.legend()
ax2.legend()

# to set title and lables we have to use set_.. instead
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

In the above plot we have already managed redundancy by not specifying title and age on both axes.<br>
But if we wanted to also remove the labels on the above plot, we can set the `sharex=` argument of `plt.subplots()` to `True`.<br>
Note: There is a similar argument for y-axis `sharey`, this would be useful if we had columns.

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,sharex=True)

# plot all dev salary
ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
# plot python salary
ax2.plot(ages, py_salaries, label='Python')
# plot javascript salary
ax2.plot(ages, js_salaries, label='JavaScript')


ax1.legend()
ax2.legend()

# to set title and lables we have to use set_.. instead
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```

Now if we wanted to create two seperate figures instead of multiple plot in the same figure. We will have to create multiple instances of `plt.subplots()`

```python
fig1,ax1=plt.subplots()
fig2,ax2=plt.subplots()

# plot all dev salary
ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')
# plot python salary
ax2.plot(ages, py_salaries, label='Python')
# plot javascript salary
ax2.plot(ages, js_salaries, label='JavaScript')


ax1.legend()
ax2.legend()

# to set title and lables we have to use set_.. instead
ax1.set_title('Median Salary (USD) by Age')
ax1.set_ylabel('Median Salary (USD)')

ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

plt.tight_layout()
plt.show()
```
