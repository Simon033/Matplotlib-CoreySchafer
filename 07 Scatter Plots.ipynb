# Scatter Plot
A scatter plot is a type of graph that shows the relationship between two variables in a data set. The data is represented by placing points on an x- and y-axis. The position of each dot on the graph indicates the values for an individual data point.

```python
# Import packages
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# (optional) set plot style
plt.style.use('seaborn-v0_8')
```

```python
# data
x = [5, 7, 8, 5, 6, 7, 9, 2, 3, 4, 4, 4, 2, 6, 3, 6, 8, 6, 4, 1]
y = [7, 4, 3, 9, 1, 3, 2, 5, 2, 4, 8, 7, 1, 6, 4, 9, 7, 7, 5, 1]
```

### A Simple Scatter Plot
To plot a scatter plot use `plt.scatter(x,y)`<br>
Scatter plot helps understand the relation between data if any.

```python
plt.scatter(x,y)
```

### Scatter Plot attributes
A scatter plot has several parameters like `s` for size of dots, `c` for color of dots, `marker` for the marker patter, etc. 

```python
plt.scatter(x,y,
           s=50,
           c='#08e8de',
           marker='*')
```

Some other commonly used and purposeful parameters are **`edgecolor`** which sets the color of the outline, **`linewidth`** which sets the outline width,**`alpha`** which sets the transparency of the dots.

```python
plt.scatter(x,y,
           s=50,
           edgecolors='#08e8de',
           linewidths=2,
           alpha=0.8)
```

### Dot colors
We can change the color of the dot based on it's value by passing in the values as the `color` parameter.

```python
plt.scatter(x,y,
            s=50,
            c=y,
            edgecolors='#08e8de',
            linewidths=2,
            alpha=0.8)
```

The same can be done using some other values as well. Lets assume that there is a satisfaction rating for each dot, we can pass that as the color parameter and achieve the same results.

```python
satisfaction=[8,5,6,8,4,7,2,6,9,0,5,2,3,6,8,4,2,8,2,8]
```

```python
plt.scatter(x,y,
            s=50,
            c=satisfaction,
            edgecolors='#08e8de',
            linewidths=2,
            alpha=0.8)
```

### Cmaps and label
We can use a variety of color range by specifying a **`cmap`** parameter.<br>
Then to plot the color bar as label create a **`plt.colorbar()`** instance to view the color bar and then set the **`cbar.set_label`** method to give a name to that label. 

Checkout the all available cmaps from the [official documentation](https://matplotlib.org/stable/users/explain/colors/colormaps.html).

```python
plt.scatter(x,y,
            s=50,
            c=satisfaction,
            cmap='cool',
            linewidths=2,
            alpha=0.8)
cbar=plt.colorbar()
cbar.set_label("Satistfaction")
```

### Dot Size
Another similar way to show properties of dots is by changing size of each dots. For eg- to represent population, sample size of that data point, etc

Similar to color we can pass in any column data to make this either x,y or some other data. Pass that data to the **`s`** parameter.

```python
# size data of each point.
sizes = [209, 486, 381, 255, 191, 315, 185, 228, 174, 538, 239, 394, 399, 153, 273, 293, 436, 501, 397, 539]
```

```python
plt.scatter(x,y,
            s=sizes,
            c=satisfaction,
            cmap='cool',
            linewidths=2,
            alpha=0.8)
cbar=plt.colorbar()
cbar.set_label("Satistfaction")
```

## Plotting Real World Data
The csv file include data about top 200 trending videos in YT from a particular day. The data includes the view counts and likes and the ratio of likes to dislikes.

```python
# import data
df=pd.read_csv('.\data\data07.csv')
df
```

```python
# create series from data
view_count = df['view_count']
likes = df['likes']
ratio = df['ratio']
```

Lets analyze the relation between the view count and likes.<br>
First lets just create simple scatter plot with two values

```python
plt.scatter(x=view_count,
            y=likes,
            edgecolors='#521efc'
           )
```

That plot has 200 dots however all of them are bundled in a corner because of an outlier in the top right.

### Log scale
In such case we can use a log scale to reduce the visual difference.<br>
To use it on an axis call the method `plt.xscale('log')` and `plt.yscale('log')` to scale the values logarithmically on both axis.

```python
plt.scatter(x=view_count,
            y=likes,
            edgecolors='#521efc'
           )

plt.xscale('log')
plt.yscale('log')

plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

Now after scaling, the outliers don't skew the visualization much.

Now lets include the ration as color, we can also use size however it might not be that visible. Although both can be used together.

```python
plt.scatter(x=view_count,
            y=likes,
            edgecolors='#521efc',
            c=ratio,
            s=ratio,
            cmap='cool',
            alpha=0.4
           )

plt.xscale('log')
plt.yscale('log')
plt.colorbar().set_label('Like to Dislike Ratio')
plt.title('Trending YouTube Videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')

plt.tight_layout()

plt.show()
```

Apart from the positive liner relation we can see that the few points that have low like to dislike ration(60-50) usually have less likes compared to other points in the same view count.
